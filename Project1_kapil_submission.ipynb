```python
%reset -f
import pandas as pd
from toolz import first,reduce
import os

CHUNK_SIZE = 500
```

```python
files = os.listdir("./Data/MinneMUDAC/")
filePaths = ['./Data/MinneMUDAC/'+ file for file in files]
filePaths.remove("./Data/MinneMUDAC/.ipynb_checkpoints")
filePaths.remove('./Data/MinneMUDAC/mces_lakes_1999_2014.txt')
filePaths
```

```python
def read_chunk(file,size):
    chunk = pd.read_csv(file,chunksize=size,sep = "|")
    return chunk
```

```python
df_1 = read_chunk(filePaths[0],CHUNK_SIZE)
first_df1 = first(df_1)
```

```python
df_2 = read_chunk(filePaths[1],CHUNK_SIZE)
first_df2 = first(df_2)
```

```python
column_same = set(first_df1.columns)== set(first_df2.columns)
column_same
```

```python
cols_common = set(first_df1.columns).intersection(set(first_df2.columns))
cols_common
```

```python
cols_differ = [set(first_df1.columns).difference(cols_common)]
cols_differ
```

```python
df_iterators = [read_chunk(filePath,CHUNK_SIZE) for filePath in filePaths]
```

```python
first_chunks = [first(df) for df in df_iterators]
```

```python
cols_set = [set(df.columns) for df in first_chunks]
cols_set
```

```python

def get_common(list_set):
     return set(reduce(lambda s1,s2:s1.intersection(s2),list_set))
```

```python
get_common(cols_set)
```
