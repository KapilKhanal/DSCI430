1.  *Stack* the columns.

![](img/survey/image1.png)

```python
cols = ['F1', 'F5', 'F2', 'F1.1', 'F2.1', 'F6', 'F4', 'F3',
       'F5.1', 'F1.2', 'F2.2', 'F6.1', 'F2.3', 'F4.1', 'F2.4', 'F5.2', 'F2.5',
       'F6.2', 'F1.3', 'F2.6', 'F5.3', 'F4.2', 'F2.7', 'F3.1', 'F2.8', 'F5.4',
       'F3.2', 'F1.4', 'F3.3', 'F1.5', 'F5.5', 'F6.3', 'F1.6', 'F5.6', 'F2.9',
       'F3.4', 'F4.3', 'F2.10', 'F1.7', 'F6.4', 'F4.4', 'F5.7', 'F3.5',
       'F2.11']
survey_stacked = (survey_data 
                 >> gather("Question","Response", cols))
survey_stacked >> head
```

2.  Read in and join `ReverseCodingItems.csv` to add a new column called `NeedsReverse` to the health survey dataframe. 

![](img/survey/image2.png)

```python
survey_join = survey_stacked >> right_join(reverse, by='Question') >> drop('Q', 'Construct', 'Qualtrics')
survey_join.head()
```

3.  Make a new column called `TempCodedValue` by recoding the `Question`s column.

![](img/survey/image3.png)

```python
scale = {'Somewhat Agree': 4 , "Strongly Agree": 5 , "Neither Agree nor Disagree": 3, "Somewhat Disagree": 2, "Strongly Disagree": 1}
survey_temp_recode = survey_join >> mutate(TempRecodeValue = X.Response.map(scale))
survey_temp_recode.head()
```

4.  Make a new column called `TempReverseValue` by recoding the `Question`s column.

![](img/survey/image4.png)

```python
reverse_scale = {1.0: 5, 2.0: 4, 3.0:3, 4.0:2, 5.0: 1 }
survey_temp_reverse = survey_temp_recode >> mutate(TempReverseValue = X.TempRecodeValue.map(reverse_scale))
survey_temp_reverse.head()
```

5.  Make a new column called `RecodedValue` that holds the correct
    value for each question based on the value in `NeedsReverse`.

![](img/survey/image5.png)

```python
from more_dfply import ifelse
```

```python
survey_new = survey_temp_reverse >> mutate(RecodedValue = ifelse(X.Needs_Reverse_Coding == 'Yes', X.TempReverseValue, X.TempRecodeValue ))
survey_new.head()
```

6.  Make a new column by *Recoding* the Question Types to *F1, F2, ..., F6. *

![](img/survey/image6.png)

```python
def get(x):
    return x.split('.')[0]
```

```python
survey_question_type = survey_new >> mutate(QuestionType = X.Question.map(get))
survey_question_type.head()
```

7.  *Aggregate* and *Unstack*.

![](img/survey/image7.png)
![](img/survey/image8.png)

```python
survey_agg = survey_question_type >> group_by(X.Participant, X.QuestionType) >> summarise(RecodedTotal = X.RecodedValue.sum())
survey_agg.head()                            
```

```python
survey_unstack = survey_agg >> spread(X.QuestionType, X.RecodedTotal) 
survey_unstack.head()
```

#### Task 3

Repackage all of your code in one pipe then write the final output to `health_survey_summary.csv`


```python
survey_all = (survey_data >> gather("Question","Response", cols) \
           >> right_join(reverse, by='Question') >> drop('Q', 'Construct', 'Qualtrics') \
           >> mutate(TempRecodeValue = X.Response.map(scale)) \
           >> mutate(TempReverseValue = X.TempRecodeValue.map(reverse_scale)) \
           >> mutate(RecodedValue = ifelse(X.Needs_Reverse_Coding == 'Yes', X.TempReverseValue, X.TempRecodeValue )) \
           >> mutate(QuestionType = X.Question.map(get))
           >> group_by(X.Participant, X.QuestionType) >> summarise(RecodedTotal = X.RecodedValue.sum())         
           >> spread(X.QuestionType, X.RecodedTotal))          
survey_all.head()
```

```python
survey_all.to_csv('./data/health_survey_summary.csv')
```
